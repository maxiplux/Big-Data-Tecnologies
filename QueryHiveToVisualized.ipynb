```python
from pyhive import hive
from wordcloud import WordCloud
import matplotlib.pyplot as plt
 
cursor = hive.connect('localhost').cursor()
cursor.execute('SELECT trending,count(1) FROM infographis GROUP BY trending')

dataset   = cursor.fetchall() 
dataset=  list( map (lambda pairs  :  [pairs[0]]  *  int (pairs[1]), dataset )  )          
dataset=sum(dataset, []) 
    
#convert list to string and generate
dataset=(" ").join(dataset)
# Create the wordcloud object
wordcloud = WordCloud(width=640, height=480,background_color ='white',margin=10,
               random_state=1).generate(dataset)
# Display the generated image:
plt.figure(figsize = (16, 16), facecolor = None) 
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.margins(x=10, y=10)
plt.show()
```
